## Chapter 1 - Machine Learning Landscape
- <b>Machine Learning</b> - is the science and art of programming computers so they can learn from data.

### Types of Machine Learning Systems

- #### Supervised Learning
    - The training data includes the desired solutions. Can be either classification(discrete) or regression(numerical) type of problem.
    - Example of applications include, classifying whether an email is spam or not. Or determining the cost of a car based on its features (# of seats, MPG, HP, tourque).
    ![supervised_classification](images/chapter_1/supervised_classification.png)
    ![supervised_regression](images/chapter_1/supervised_regression.png)
    - The predictors or features are the columns which can help in determining what the label should be.
    - Important supervised Algorithims
        - K-Nearest Neighbours
        - Linear Regression
        - Logistic Regression
        - Support Vector Machines (SVM)
        - Decision Tress
        - Random Forests
        - Neural Networks
- #### Unsupervised Learning
    - The training data does not include the "answers" - it is not labelled. The system tries to learn without a teacher.
    - Example of application include, clustering similar users that visit a blog. In this case we don't know which users are similar, but the clustering algorithim will group together users which have similar features (ex. Male, like comics, time of visit to blogsite.)
    - Important unsupervised Algorithms
        - Clustering
            - K-means
            - DBSCAN
            - Hierarchical Cluster Analysis (HCA)
        - Anomaly detection and novelty detection
            - One class SVM
            - Isolation Forest
        - Visualization and Dimensionality Reduction
            - Principal Component Analysis (PCA)
            - Kernel PCA
            - Local-linear Embedding (LLE)
            - t-distributed Stochastic Neighbor Embedding (t-SNE)
        - Association rule Learning
            - Apriori
            - Eclat
            
- #### Semisupervised Learning
    - When the data is partially labelled, semisupervised algorithims are utilized.
    - Example of applications include a photo tagging service, which may need help indentifying the names of the individual in the begining but then begin to label by themselves.
- #### Reinforcement Learning
    - The learning system is called the Agent.
    - The agent can observe the environment, select and perform actions, and get rewards or penalties in response.
    - Using the rewards and penalties, it can come up with a best strategy called a policy. Essentially the policy will look to maximize the rewards.
    ![reinforcement_learning](images/chapter_1/reinforcement_learning.png)

### Batch and Online Learning

- #### Batch Learning
    - The system is incapable of learning incrementally. It must be trained using all availaible data.
    - Takes a long time depending on the size of the data.
    - Generally done offline. 
    - System is trained, launched into production. If new data comes in, then the system needs to be taken offline, re-trained using all data again (including the new data) and then relaunced into production.

- #### Online Learning
    - The system is trained incrementally by being exposed to data in smaller chunks. The chunks can be individual or in small chunks called mini-batches. 
    - Becuase the training data is smaller when compared to the entire dataset, the training set is a lot faster.
    - Requires less computer resources.

### Instance-Based Versus Model-Based Learning
- #### Instance-based learning
    - The model compares the new value to a existing value it has already seen before in the training set.
    - The new value will be classified the same as the closest value it has already seen.
    - This can also work for regression problems, by taking the average of the closest values from the training dataset.
    - Examples include K nearest neighbours.

- #### Model-based learning
    - A model is created where the parameters of the new value can be entered into the system and a resulting answer will generate.
    - Example include using linear regression to predict house prices. The linear regression model in this case will generate an equation in the form $y =mx+b$. 
    

### Overfitting vs Underfitting
- #### Overfitting
    - The model performs very well on the training data, but fails to generalize to the test data.
    - Can use regularization techniques to punish overfitting.

- #### Underfitting
    - Model is too simple and fails to learn the underlying structure of the data.
    - Example, if we are trying to fit a linear regression mode to a model which should actually be better represented by a polynomial.
